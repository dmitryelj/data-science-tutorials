```python
import os
import pandas as pd
import numpy as np

from bokeh.io import show, output_notebook, export_png
from bokeh.plotting import figure, output_file
from bokeh.models import ColumnDataSource, LabelSet
output_notebook()
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline


def show_figure(w: int, h: int, fig: figure):
    """ Matplotlib wrapper for displaying Bokeh in GitHub """
    plt.rcParams['figure.figsize'] = [w, h]

    png_file = export_png(fig, filename="temp_image.png")
    plt.imshow(mpimg.imread(png_file))
    os.remove(png_file)
```

# Load

```python
# URL: https://opendata.rdw.nl/Voertuigen/Open-Data-RDW-Gekentekende_voertuigen/m9d7-ebf2

filename = "Open_Data_RDW__Gekentekende_voertuigen.csv"
# df = pd.read_csv(filename)
# display(df)

# Read first lines
with open(filename, 'r') as f:
    header_str = f.readline()
    print(header_str)
    for _ in range(2):
        print(f.readline())
```

```python
# Specify columns to load
cols = ['Kenteken', 'Voertuigsoort', 'Merk', 'Handelsbenaming', 'Eerste kleur', 'Tweede kleur', 'Catalogusprijs']
cols_date = []
for c in header_str.split(","):
    if "datum" in c.lower() and 'vervaldatum' not in c.lower() and 'bpm' not in c.lower():
        cols_date += [c]
print(cols)
print(cols_date)
```

```python
filename = "Open_Data_RDW__Gekentekende_voertuigen_short.csv"
df = pd.read_csv(filename, usecols=cols + cols_date, parse_dates=cols_date, dtype={"Catalogusprijs": pd.UInt32Dtype()})
display(df)
df.info(memory_usage="deep")

# df1 = df[df['Voertuigsoort'] == "Personenauto"]
# display(df1)
# df1.to_csv(filename.replace(".csv", "_short.csv"), sep=',', encoding='utf-8')
# print("File saved")

# del df1

# df.rename(translations_dict_en_nl, axis='columns', inplace=True)
# display(df)

# df3 = pd.read_csv(filename, usecols=cols + cols_date, parse_dates=cols_date)
# display(df3)

# df.info(memory_usage="deep")
```

```python
translations_dict_en_nl = {
                           'Kenteken': 'License plate', 'Voertuigsoort': 'Type', 'Merk': 'Model', 'Handelsbenaming': 'Trade name', 'Eerste kleur': 'First color', 'Tweede kleur': 'Second color', 'Catalogusprijs': 'Catalog price',
                           'Datum eerste toelating': 'First admission', 
                           'Datum eerste toelating DT': 'First admission DT', 
                           'Datum tenaamstelling': 'First registration', 
                           'Datum tenaamstelling DT': 'First registration DT',
                           'Datum eerste tenaamstelling in Nederland': 'First registration NL', 
                           'Datum eerste tenaamstelling in Nederland DT': 'First registration NL DT', 
                           'Registratie datum goedkeuring (afschrijvingsmoment BPM)': 'Registration date approval (depreciation moment BPM)',
                           'Registratie datum goedkeuring (afschrijvingsmoment BPM) DT': 'Registration date approval (depreciation moment BPM) DT'
                          }

df.rename(translations_dict_en_nl, axis='columns', inplace=True)
display(df)
```

```python
# Calculate percentiles

prices = df[df['Catalog price'].notna()]['Catalog price'].to_numpy()

print("Mean:", np.mean(prices))
print("Standard deviation:", np.std(prices))
print("Perceitiles [25, 50, 75, 95]:", np.percentile(prices, [25, 50, 75, 95]))

del prices
```

```python
df_price = df[df['Catalog price'].notna()]

# p_all = df_price['Catalog price'].shape[0]
# p1 = df_price.query('`Catalog price` < 100000')['Catalog price'].shape[0]
# p2 = df_price.query('`Catalog price` >= 100000')['Catalog price'].shape[0]
# print("P1:", p1, p2, 100*p1/p_all, 100*p2/p_all)

prices = df_price.query('`Catalog price` < 100000')['Catalog price'].to_numpy()
# prices = np.array([10, 12, 13, 18, 18, 19, 20, 25, 28, 30, 31, 32, 40])
print("Cars total", df_price.shape[0])
print("Cars less than price:", prices.shape[0], prices.shape[0]*100/df_price.shape[0])
# display(prices[:10])
# display(prices[-10:])

p_50 = np.percentile(prices, 50)
p_95 = np.percentile(prices, 95)
hist, edges = np.histogram(prices, density=False, bins=50)

del prices
del df_price


p = figure(width=1400, height=500, title="Car prices distribution in the Netherlands")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")
p.ray(x=p_50, y=0, length=0, angle=np.pi/2, line_width=2, color="blue", alpha=0.5)
p.ray(x=p_95, y=0, length=0, angle=np.pi/2, line_width=2, color="blue", alpha=0.5)
labels_source = ColumnDataSource(data={"x_labels": [p_50, p_95], "y_labels": [0, 0], "names": ["Median", "95% percentile"]})
p.add_layout(LabelSet(x="x_labels", y="y_labels", text="names", x_offset=5, y_offset=5, source=labels_source))
p.x_range.start = 0
p.y_range.start = 0
p.xaxis.axis_label = "Price, EUR"
p.yaxis.axis_label = "Amount"
p.left[0].formatter.use_scientific = False
p.below[0].formatter.use_scientific = False
show(p)

# Matplotlib Preview (GitHub only)
# show_figure(w=14, h=7, fig=p)
```
